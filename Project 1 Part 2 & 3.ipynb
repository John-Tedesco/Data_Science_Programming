```python
for k,v in referral_dict.items():
    df.rename(columns=({f"How did you hear about County College of Morris? [{k}]":f"referral [{v}]"}),
              inplace=True)
```

```python
for k,v in decision_impact_dict.items():
    df.rename(columns=({f"To what extent did the following impact your decision to attend County College of Morris? [{k}]":f"decision_impact [{v}]"}),
              inplace=True)
```

```python
for k,v in attended_event_dict.items():
    df.rename(columns=({f"Prior to applying to college, did you participate in any of the following events or activities at the County College of Morris and/or with the Department of Information Technologies, if at all? [{k}]":f"attended_event [{v}]"}),
             inplace=True)
```

```python
for k,v in experience_impact_dict.items():
    df.rename(columns=({f"To what extent did the following activities or experience impact your decision to enroll in an computing course at CCM? [{k}]":f"experience_impact [{v}]"}),
             inplace=True)
```

```python
for k,v in motivation_dict.items():
    df.rename(columns=({f"What motivated you to seek a computing degree/certificate at CCM?  [{k}]":f"motivation [{v}]"}),
             inplace=True)
```

```python
for k,v in info_source_dict.items():
    df.rename(columns=({f"Did you receive information about the CCM computing programs from any of the following sources? [{k}]":f"info_source [{v}]"}),
             inplace=True)
```

```python
course_list = ["CMP 120","CMP 128","CMP 130","CMP 131","CMP 239"]
for i in course_list:
    df.loc[df["enrolled_course"].str.contains(i, na="NotAvailable"), "enrolled_course"] = i
```

```python
for v in referral_dict.values():
    df.loc[df[f"referral [{v}]"].str.contains("yes", case=False, na=False), f"referral [{v}]"] = "yes"
    df.loc[df[f"referral [{v}]"].str.contains("no", case=False, na=False), f"referral [{v}]"] = "no"
    df.loc[df[f"referral [{v}]"].str.contains("recall", case=False, na=False), f"referral [{v}]"] = "unsure"
```

```python
for v in decision_impact_dict.values():
    df.loc[df[f"decision_impact [{v}]"].str.contains("N/A", case=False, na=False), f"decision_impact [{v}]"] = "Na"
    df.loc[df[f"decision_impact [{v}]"].str.contains("no", case=False, na=False), f"decision_impact [{v}]"] = "none"
    df.loc[df[f"decision_impact [{v}]"].str.contains("some", case=False, na=False), f"decision_impact [{v}]"] = "some"
    df.loc[df[f"decision_impact [{v}]"].str.contains("high", case=False, na=False), f"decision_impact [{v}]"] = "high"
```

```python
for v in attended_event_dict.values():
    df.loc[df[f"attended_event [{v}]"].str.contains("yes", case=False, na=False), f"attended_event [{v}]"] = "yes"
    df.loc[df[f"attended_event [{v}]"].str.contains("no", case=False, na=False), f"attended_event [{v}]"] = "no"
    df.loc[df[f"attended_event [{v}]"].str.contains("sure", case=False, na=False), f"attended_event [{v}]"] = "unsure"
```

```python
for v in experience_impact_dict.values():
    df.loc[df[f"experience_impact [{v}]"].str.contains("N/A", case=False, na=False), f"experience_impact [{v}]"] = "Na"
    df.loc[df[f"experience_impact [{v}]"].str.contains("no", case=False, na=False), f"experience_impact [{v}]"] = "none"
    df.loc[df[f"experience_impact [{v}]"].str.contains("some", case=False, na=False), f"experience_impact [{v}]"] = "some"
    df.loc[df[f"experience_impact [{v}]"].str.contains("high", case=False, na=False), f"experience_impact [{v}]"] = "high"
```

```python
for v in motivation_dict.values():
    df.loc[df[f"motivation [{v}]"].str.contains("yes", case=False, na=False), f"motivation [{v}]"] = "yes"
    df.loc[df[f"motivation [{v}]"].str.contains("no", case=False, na=False), f"motivation [{v}]"] = "no"
```

```python
for v in info_source_dict.values():
    df.loc[df[f"info_source [{v}]"].str.contains("yes", case=False, na=False), f"info_source [{v}]"] = "yes"
    df.loc[df[f"info_source [{v}]"].str.contains("no", case=False, na=False), f"info_source [{v}]"] = "no"
    df.loc[df[f"info_source [{v}]"].str.contains("recall", case=False, na=False), f"info_source [{v}]"] = "unsure"
```

```python
gender_dict = {"Woman":"w", "Man":"m", "Non-binary":"non-b", "I do not identify":"not_id", "Prefer not to say":"not_say"}
for k,v in gender_dict.items():
    df.loc[df["gender"].str.contains(k, case=False, na=False), "gender"] = v
```

```python
race_dict = {"Hispanic or Latino":"latino", "American Indian/Native American/Alaska Native":"native", "Asian":"asian",
             "Black/African American":"black", "Native Hawaiian/Other Pacific Islander":"islander", "White/Caucasian":"white",
             "Multi-Racial":"multi_race", "Choose not to reply":"no_reply"}
for k,v in race_dict.items():
    #df.loc[df["race/ethnicity"].str.contains(k, case=False, na=False), "race/ethnicity"] = v
    df["race/ethnicity"] = df["race/ethnicity"].replace(k,v)
df.loc[df["race/ethnicity"].str.contains(";", case=False, na=False), "race/ethnicity"] = "multi_race"
df.loc[df["age "].str.contains("18", case=False, na=False), "age "] = "< 18"
```

```python
computing_majors = ["comp_sci", "IT", "game_dev", "digital_media", "tech_studies", "info_sec", "data_analytics", "first"]
# ChatGPT did help with the following
df["computing_major_first"] = df["major_choice/change"].apply(lambda i: "comp_major" if i in computing_majors else "other")
majors_count = df["computing_major_first"].value_counts()
```
